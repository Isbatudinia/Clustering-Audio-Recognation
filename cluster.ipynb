```python
import os
import librosa
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
folder_path = r"C:\Users\USER\Documents\jurnal pakmaman\campur"
```

```python
wav_files = [f for f in os.listdir(folder_path) if f.endswith('.wav')]
```

```python
data = []
```

```python
for wav_file in wav_files:
    file_path = os.path.join(folder_path, wav_file)
    
    # Membaca file WAV
    audio, sample_rate = librosa.load(file_path, sr=None)

    # Normalisasi
    audio = audio / np.max(np.abs(audio))

    # Penghapusan Noise (Contoh sederhana dengan trimming)
    audio_trimmed, _ = librosa.effects.trim(audio)

    # Ekstraksi fitur (contoh: MFCC)
    mfccs = librosa.feature.mfcc(y=audio_trimmed, sr=sample_rate, n_mfcc=13)

    # Mengambil mean dari setiap MFCC untuk menyederhanakan (opsional)
    mfccs_mean = np.mean(mfccs.T, axis=0)

    # Simpan nama file dan fitur MFCC
    row = [wav_file] + mfccs_mean.tolist()
    data.append(row)
```

```python
columns = ['file_name'] + [f'mfcc_{i}' for i in range(1, 14)]  # MFCC 1 hingga MFCC 13
df = pd.DataFrame(data, columns=columns)

# Menyimpan ke CSV
df.to_csv('extract.csv', index=False)

print("Ekstraksi selesai dan disimpan ke extract.csv")
```

```python
datac = data.drop(columns=['file_name'])
```

```python
X = np.array(datac)
```

```python
n_clusters = 2
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
```

```python
kmeans.fit(X)
```

```python
labels = kmeans.labels_
```

```python
datac['label'] = labels
```

```python
datac.to_csv('dataclusters.csv', index=False)
```
